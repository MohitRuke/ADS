```python
X = data[['x','y','z']]
Y = data['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
metrics.explained_variance_score(y_test, predictions)
```

```python
a = data[['carat','y']]
b = data['price']
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.4, random_state=101)
```

```python
lm1 = LinearRegression()
```

```python
lm1.fit(a_train,b_train)
```

```python
print(lm1.intercept_)
```

```python
coeff_df1 = pd.DataFrame(lm1.coef_,a.columns,columns=['Coefficient'])
coeff_df1
```

```python
predictions = lm1.predict(a_test)
```

```python
plt.scatter(b_test,predictions)
```

```python
print('MAE:', metrics.mean_absolute_error(b_test, predictions))
print('MSE:', metrics.mean_squared_error(b_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(b_test, predictions)))
```

```python
metrics.explained_variance_score(b_test, predictions)
```
